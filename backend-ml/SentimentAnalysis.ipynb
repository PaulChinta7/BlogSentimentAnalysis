```python
import pandas as pd

file_path = 'Twitter_Data.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
df.shape
```

```python
# Check null values
cleantext_null_values=df['clean_text'].isnull().sum()
category_null_values=df['category'].isnull().sum()
print("null values in clean-text : "+str(cleantext_null_values))
print("null values in category : "+str(category_null_values))
```

```python
# 0-, 1-positive, -1 negative
clean_df=df.dropna()
pd.set_option('display.max_colwidth', None)
clean_df.shape
```

```python
clean_df[:2]
```

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
clean_df['clean_text'] = clean_df['clean_text'].apply(lambda tokens: ' '.join(tokens))
```

```python
clean_df.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
import nltk
```

```python
vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(clean_df['clean_text'])

y = clean_df['category']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred, target_names=['Negative', 'Neutral', 'Positive']))
print(f'Accuracy: {accuracy_score(y_test, y_pred)}')
```

```python
import pickle


with open('model.pkl', 'wb') as model_file:
    pickle.dump(model, model_file)

with open('vectorizer.pkl', 'wb') as vectorizer_file:
    pickle.dump(vectorizer, vectorizer_file)
```
